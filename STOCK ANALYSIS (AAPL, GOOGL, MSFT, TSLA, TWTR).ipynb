# Financial Time Series - Return and Risk

```python
import numpy as np
```

```python
norm.plot(figsize=(15,8), fontsize=13)
plt.legend(fontsize = 13);
```

```python
close.head()
```

```python
close.pct_change().dropna()
```

```python
ret = close.pct_change().dropna()
```

```python
ret.head()
```

```python
ret.describe().T.loc[:,['mean','std']]
```

```python
summary = ret.describe().T.loc[:,['mean','std']]
summary
```

```python
summary['mean'] =summary['mean'].mul(252)
summary['std'] = summary['std'].mul(np.sqrt(252))
```

```python
summary
```

```python
summary.plot(kind='scatter', x= 'std', y='mean', figsize=(15,10), s= 200,fontsize=15)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, 'std']+0.004, summary.loc[i,'mean']+0.003), size=15)
plt.xlabel('ann. Risk(std)', fontsize=15)
plt.ylabel('ann. Return', fontsize=15)
plt.title('RISK | RETURN', fontsize=20);
```
